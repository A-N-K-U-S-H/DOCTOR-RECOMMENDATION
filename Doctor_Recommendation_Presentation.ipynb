```python
df = df.drop('Overall_rating',axis=1)
df
```

```python
specs_values = list(df['Specialization'].drop_duplicates().values)
specs_values
```

```python
# Create a dictionary to store doctor ratings
doctor_ratings = {doctor: rating for doctor, rating in zip(df["Doctor_name"], df["Rating"])}

# Create a dictionary to store patient counts
patient_counts = {doctor: count for doctor, count in zip(df["Doctor_name"], df["Patient_count"])}
```

```python
# Calculate cosine similarity between doctor ratings
doctor_names = df["Doctor_name"].tolist()
ratings = [doctor_ratings[doc] for doc in doctor_names]
ratings_matrix = [ratings] * len(ratings)  # Create a matrix of ratings
cosine_sim_matrix = cosine_similarity(ratings_matrix)
```

```python
# Normalize patient counts and ratings
max_patient_count = max(patient_counts.values())
max_rating = max(doctor_ratings.values())

for doctor in doctor_ratings:
    patient_counts[doctor] /= max_patient_count
    doctor_ratings[doctor] /= max_rating
```

```python
# Calculate trust factor for each doctor
trust_factors = {}

for i, doctor in enumerate(doctor_ratings):
    similarity_weight = 0.1
    patient_count_weight = 0.3
    rating_weight = 0.6

    # Calculate the sum of cosine similarities with other doctors
    similarity_sum = sum(cosine_sim_matrix[i])

    trust_factor = (
        (similarity_weight * similarity_sum +
        patient_count_weight * patient_counts[doctor] +
        rating_weight * doctor_ratings[doctor]) / 10
    )
    trust_factors[doctor] = trust_factor

# Print the trust factor for each doctor
for doctor, trust_factor in trust_factors.items():
    print(f"Trust factor for {doctor}: {trust_factor:.4f}")
```

```python
# Add the Trust Factor column to the DataFrame
df["Trust Factor"] = df["Doctor_name"].map(trust_factors)
df
```

```python
# Data transformation (scaling numeric features)
scaler = StandardScaler()
df['Trust Factor'] = scaler.fit_transform(df['Trust Factor'].values.reshape(-1, 1))
```

```python
# Structuring data for collaborative filtering
user_doctor_matrix = df.pivot_table(index='User_ID', columns='Doctor_ID', values='Trust Factor', fill_value=0)
matrix_shape = user_doctor_matrix.shape
num_doctors = matrix_shape[0] # Number of doctors
num_users = matrix_shape[1] # Number of users

# Verify the indexing (should be 0 to N-1)
doctor_ids = range(num_doctors) # Create a list of doctor IDs (0 to N-1)
user_ids = range(num_users) # Create a list of user IDs (0 to N-1)

# Print the number of doctors, users, and the shape of the matrix
print(f"Number of doctors: {num_doctors}")
print(f"Number of users: {num_users}")
print(f"Matrix shape: {matrix_shape}")
print(f"Doctor IDs: {doctor_ids}")
print(f"User IDs: {user_ids}")
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Data splitting
train_data, test_data = train_test_split(user_doctor_matrix, test_size=0.2, random_state=42)
```

```python
# Calculate doctor similarities using cosine similarity
doctor_similarities = cosine_similarity(user_doctor_matrix)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
# Create a K-nearest neighbors model
k = 3
knn = NearestNeighbors(n_neighbors=k, metric='cosine')
knn.fit(doctor_similarities)

def get_top_recommendations(doctor_id, num_recommendations=k):
    doctor_idx = doctor_id - 101
    doctor_specialization = df[df['Doctor_ID'] == doctor_id]['Specialization'].values[0]

    similar_doctors_with_specialization = df[(df['Specialization'] == doctor_specialization) & (df['Doctor_ID'] != doctor_id)]

    if len(similar_doctors_with_specialization) >= k:
        similar_doctors = knn.kneighbors([doctor_similarities[doctor_idx]], n_neighbors=num_recommendations+1)
        similar_doctor_indices = similar_doctors[1][0][1:]
        recommended_doctors = [(doctor_ids[idx] + 101) for idx in similar_doctor_indices]
    else:
        similar_doctors_sorted = similar_doctors_with_specialization.copy()
        similar_doctors_sorted = similar_doctors_sorted.sort_values(by='Trust_Factor_Score', ascending=False)
        recommended_doctors = similar_doctors_sorted['Doctor_ID'].tolist()

    return recommended_doctors[:num_recommendations]
```

```python
# Get recommendations for a specific doctor
doctor_id = 137
recommendations = get_top_recommendations(doctor_id)
print(f"Top recommendations for Doctor_ID {doctor_id}: {recommendations}")
```
